```python
import requests as req
import pandas as pd
# Google developer API key
from config_local import gkey
```

```python
# read sites 
xls = pd.ExcelFile('Project1_AmazonSites.xlsx') 
sites_df=xls.parse('AmazonSites', dtype=str) 
```

```python
components= sites_df[['Site Name','State','Zip Code','Amazon City','Latitude','Longitude']]
components.head()
```

```python
# Use google search to get nearby schools and their ratings
```

```python
# geocoordinates
target_coordinates = ''
target_search = "Hospitals"
target_radius = 8000
target_type = "Hospitals"
# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
import time

nearby = []
ammenities = ['Hospital','Restaurant','Theatre','Shopping']
for a in ammenities:
    params['keyword'] = a
    params['type'] = a
    
    for index, row in components.iterrows():
        lat = row['Latitude']
        lng = row['Longitude']
        site = row['Site Name']
        coordinates = str(lat)+','+ str(lng)
        params['location'] = coordinates
        # run a request using our params dictionary
        response = req.get(base_url, params=params)
        print(response.url)
        # convert response to json
        data = response.json()
        number = len(data["results"])
        print(number)
        # Print the name and address of the first restaurant that appears
        for x in range(number):
            if "rating" in data["results"][x].keys():
                rating = data["results"][x]["rating"]
            else:
                rating = 'NAN'
            nearby.append({"Site Name":site,
                           "count of":number,
                        "Latitude":lat,
                        "Longitude":lng,
                        "Name":data["results"][x]["name"],
                        "Rating":rating})    
    time.sleep(30)    
```

```python
ammenities_df = pd.DataFrame(nearby)
ammenities_df.to_csv("Ammenities.csv")
```
